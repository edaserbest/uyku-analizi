```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data/student_sleep_patterns.csv")

df.head()
```

```python
df.info()
```

Yorum: Veri setinde toplam 500 gözlem ve 14 değişken bulunmaktadır. Değişken tipleri incelenmiştir.

```python
df.isnull().sum()
```

**Yorum:** Veri setinde eksik değer bulunmamaktadır. Bu nedenle eksik veri temizleme işlemi yapılmamıştır.

```python
df["Gender"].value_counts()
```

Yorum: Cinsiyet değişkenindeki öğrenci sayıları incelenmiştir.

```python
df["Sleep_Quality"].mean()
```

Yorum: Öğrencilerin ortalama uyku kalitesi hesaplanmıştır.

```python
korelasyon = df["Study_Hours"].corr(df["Sleep_Duration"])

print("Korelasyon:", korelasyon)
```

Yorum: Çalışma süresi ile uyku süresi arasındaki korelasyon değeri -0.011 bulunmuştur. Bu değer sıfıra yakın olduğundan güçlü bir ilişki gözlenmemiştir.


```python
df.groupby("Gender")["Sleep_Duration"].mean()
```

Yorum: Cinsiyet gruplarına göre ortalama uyku süreleri birbirine yakın bulunmuştur.


```python
df.describe()
```

Yorum: Sayısal değişkenlerin ortalama, standart sapma, minimum ve maksimum değerleri incelenmiştir. Öğrencilerin ortalama uyku süresi yaklaşık 6.47 saat olarak bulunmuştur.

```python
df.groupby("University_Year")["Sleep_Duration"].mean()
```

Yorum: Sınıf düzeylerine göre ortalama uyku süreleri incelenmiştir. Sınıflar arasında büyük farklılık gözlenmemiştir.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sleep_Duration"], bins=10)

plt.title("Uyku Süresi Dağılımı")
plt.xlabel("Uyku Süresi (Saat)")
plt.ylabel("Öğrenci Sayısı")

plt.show()
```

**Yorum:** Öğrencilerin uyku süreleri yaklaşık 4 ile 9 saat arasında değişmektedir. Uyku sürelerinin orta değerlerde yoğunlaştığı görülmektedir.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Study_Hours"], df["Sleep_Duration"])

plt.title("Çalışma Süresi ve Uyku Süresi İlişkisi")
plt.xlabel("Çalışma Süresi (Saat)")
plt.ylabel("Uyku Süresi (Saat)")

plt.show()
```

**Yorum:** Çalışma süresi ile uyku süresi arasında belirgin bir doğrusal ilişki gözlenmemiştir. Noktalar grafik üzerinde dağınık yapı göstermektedir.

```python
ortalama_uyku = df.groupby("Gender")["Sleep_Duration"].mean()

plt.figure(figsize=(7,5))

ortalama_uyku.plot(kind="bar")

plt.title("Cinsiyete Göre Ortalama Uyku Süresi")
plt.xlabel("Cinsiyet")
plt.ylabel("Ortalama Uyku Süresi (Saat)")

plt.show()
```

**Yorum:** Cinsiyet gruplarına göre ortalama uyku süreleri birbirine yakın bulunmuştur. Büyük bir farklılık gözlenmemiştir.

```python
plt.figure(figsize=(8,5))

plt.boxplot(df["Sleep_Duration"])

plt.title("Uyku Süresi Kutu Grafiği")
plt.ylabel("Uyku Süresi (Saat)")

plt.show()
```

**Yorum:** Uyku sürelerinin medyan değeri yaklaşık 6.5 saat civarındadır. Belirgin aykırı değer gözlenmemiştir.

```python
plt.figure(figsize=(8,5))

df["Sleep_Quality"].value_counts().sort_index().plot(kind="bar")

plt.title("Uyku Kalitesi Dağılımı")
plt.xlabel("Uyku Kalitesi")
plt.ylabel("Öğrenci Sayısı")

plt.show()
```

**Yorum:** Uyku kalitesi farklı seviyelerde dağılmıştır. Öğrenciler arasında değişkenlik olduğu görülmektedir.
